Filtering Data

And &

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

Or |

```python
df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
```

Filtering and Conditionals

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

Reset Index

```python
# new_df = new_df.reset_index()

#if you want to delete the index col
new_df.reset_index(drop=True, inplace=True)
new_df
```

Filter using strings

```python
#look for names containing mega
df.loc[df['Name'].str.contains('Mega')]
df.head()
```

```python
#look for names without mega
df.loc[~df['Name'].str.contains('Mega')]
```

Regex filtering

```python
# flags=re.I means ignore case
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
#get all pokemon names that starts with pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
```
